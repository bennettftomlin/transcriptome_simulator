```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from matplotlib.ticker import FuncFormatter
```

```python
results = pd.read_csv("results.csv")
original = pd.read_csv("FPMK_key.csv")
p_value = pd.read_csv("p_value.csv")
```

```python
results['mean'] = results.mean(axis=1)
results['log_mean_sq'] = (np.log2(results['mean'])**2)
```

```python
original['mean'] = original.mean(axis=1)
original['log_mean_sq'] = (np.log2(original['mean'])**2)
```

```python
original['mean'].describe()
```

```python
plt.style.available
plt.style.use('tableau-colorblind10')
```

```python
fig,ax0 = plt.subplots(nrows=1, ncols=1, sharey=True, sharex=False, figsize=(18, 12), dpi=300)
ax0.set_xlim(0.001,1)
#ax1.set_xlim(0.1,0.5)
plt.hist(results['mean'], histtype='step', bins=250, range=(0.001,1), label="Simulated Results")
plt.hist(original['mean'], histtype='step', bins=250, range=(0.001,1), label="Pilot Reads")
fig.suptitle(t="The Average Expression of Original Pilot Reads and Simulated Reads", ha='center', va='top', size=24, weight=2)
plt.xlabel("Average Expression", size=18, weight=1)
plt.ylabel("Number of Genes", size=18, weight=1)
ax0.tick_params(labelsize=14)
plt.legend(prop={'size': 16})
fig.savefig("AverageExpression.png")
```

```python
fig,ax0 = plt.subplots(figsize=(16, 16), dpi=300)
plt.scatter(x=results['log_mean_sq'], y=original['log_mean_sq'], s=2, alpha=0.8)

fig.suptitle(t="The Squared Log Expression of Original Pilot Reads and Simulated Reads", ha='center', va='top', size=22, weight=2)
plt.xlabel("Simulated Expression", size=18, weight=1)
plt.ylabel("Original Expression", size=18, weight=1)
ax0.tick_params(labelsize=14)
fig.savefig("LogExpression.png")
```

```python
fig,ax0 = plt.subplots(nrows=1, ncols=1, sharey=True, sharex=False, figsize=(20, 14), dpi=300)
ax0.set_xlim(0,1)

plt.hist(p_value['p_value'], histtype='bar', bins=100, range=(0,1), label="Simulated Results")
fig.suptitle(t="P Values Resulting from Comparison Between Simulated and Pilot Distributions", ha='center', va='top', size=24, weight=2)
plt.xlabel("P Values of Distribution", size=18, weight=1)
plt.ylabel("Number of Genes", size=18, weight=1)
ax0.tick_params(labelsize=14)
fig.savefig("PValuePlot.png")
```
